```python
from liteparse import LiteParse

parser = LiteParse()

# result = parser.parse(
#     "/Users/r4yen/Desktop/Research/semi-formal/repo/semipy-package/examples/data/receipts/images/038.jpg",
#     ocr_enabled=True,
# )

result = parser.batch_parse(
    input_dir="/Users/r4yen/Desktop/Research/semi-formal/repo/semipy-package/examples/data/receipts/images",
    output_dir="/Users/r4yen/Desktop/Research/semi-formal/repo/semipy-package/examples/output/receipts",
    recursive=True,              # Include subdirectories
    extension_filter=".jpg",     # Only PDF files
    ocr_enabled=True,
)

print(f"Parsed {result.success_count} files in {result.total_time_seconds}s")
print(f"Average: {result.avg_time_ms}ms per file")
```

```python
from llama_cloud import AsyncLlamaCloud
from dotenv import load_dotenv
load_dotenv()
import os

client = AsyncLlamaCloud(api_key=os.getenv("LLAMA_CLOUD_API_KEY"))

file_obj = await client.files.create(
    file="/Users/r4yen/Desktop/Research/semi-formal/repo/semipy-package/examples/data/receipts/images/038.jpg",
    purpose="parse"
)

result = await client.parsing.parse(
    file_id=file_obj.id,
    # The parsing tier. Options: fast, cost_effective, agentic, agentic_plus,
    tier="agentic",
    # The version of the parsing tier to use. Use 'latest' for the most recent version,
    version="latest",
    # 'expand' controls which result fields are returned in the response.,
    # Without it, only job metadata is returned. Common fields:,
    # - "markdown_full", "text_full": Full document content,
    # - "markdown", "text", "items": Page-level content,
    # - "images_content_metadata": Presigned URLs for images,
    expand=["markdown_full", "text_full"],
)

# Access the full document content
print("Full markdown:")
print(result.markdown_full)

print("Full text:")
print(result.text_full)
```
