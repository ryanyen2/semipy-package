```python
import time
from collections import deque

from rich.live import Live
from rich.panel import Panel
from rich.syntax import Syntax

VISIBLE_LINES = 2
lines = deque([""], maxlen=VISIBLE_LINES)

def push_chunk(chunk: str) -> None:
    parts = chunk.split("\n")

    # append to current last line
    lines[-1] += parts[0]

    # every newline starts a new visible line
    for part in parts[1:]:
        lines.append(part)

def render():
    visible = list(lines)

    # pad so the panel keeps a stable height
    if len(visible) < VISIBLE_LINES:
        visible += [""] * (VISIBLE_LINES - len(visible))

    code = "\n".join(visible)

    return Panel(
        Syntax(code, "python", line_numbers=False, word_wrap=False),
        title="Generating",
        subtitle=f"last {VISIBLE_LINES} lines",
        border_style="cyan",
    )

# demo stream
chunks = [
    "def fib(n):\n",
    "    if n <= 1:\n",
    "        return n\n",
    "\n",
    "    a, b = 0, 1\n",
    "    for _ in range(n - 1):\n",
    "        a, b = b, a + b\n",
    "    return b\n",
]

with Live(
    render(),
    auto_refresh=False,
    refresh_per_second=12,
    vertical_overflow="crop",
) as live:
    for chunk in chunks:
        push_chunk(chunk)
        live.update(render(), refresh=True)
        time.sleep(0.2)
```
