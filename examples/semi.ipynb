```python
from semipy import semiformal, semi
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/covid_19_clean_complete.csv")
df = df.sample(100)
df.head()
```

```python
df['Continent'] = df['Country/Region'].apply(lambda x: semi(f'{x}s continent'))
df['Continent'].value_counts()
```

```python
df['Country'] = df['Country/Region']
df['Continent2'] = df['Country'].apply(lambda x: semi(f'{x}s continent'))
df['Continent2'].value_counts()
```
